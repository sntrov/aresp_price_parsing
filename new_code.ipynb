Загружаем датафрейм с сайта ares-met.ru  и приводим его к правильному виду:

```python
df = pd.read_excel('http://ares-met.ru/price-list.xls', header=4)
display(df.head(5))
```

```python
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='any')
df = df.rename(
    columns={
        '1.Сталь углеродистая, легированная и высоколегированная': 'name',
        'Unnamed: 12' : 'price',
        'Unnamed: 14' : 'count'
    })
df= df.drop(['Unnamed: 13'], axis=1)
display(df.head(10))
```
